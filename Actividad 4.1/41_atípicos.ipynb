```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
# Antes del tratamiento de outliers, hay que tratar los nulos
# Solo hay nulos en nombre_vendedor
df.isnull().sum()
```

```python
# Como es un string, se reemplaza por 'Desconocido'
df['NOMBRE_VENDEDOR'] = df['NOMBRE_VENDEDOR'].fillna('Desconocido')
df.isnull().sum()
```

```python
# Veamos de qué tipo son las variables
df.info()
```

```python
# Separamos las variables cualitativas de las cuantitativas
cualitativas = df.select_dtypes(include=['object', 'datetime64'])
cuantitativas = df.select_dtypes(include=['float64', 'int64'])
```

```python
# Verificamos que se hayan separado correctamente
print(cualitativas.columns)
print(cuantitativas.columns)
```

```python
# Sacamos un diagrama e cajas
fig = plt.figure(figsize=(30, 30))
cuantitativas.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

## Procedimiento "desviación estándar" para eliminar outliers en df

```python
y = cuantitativas
limite_superior = y.mean() + 3 * y.std()
limite_inferior = y.mean() - 3 * y.std()
print("Límite superior permitido: ",limite_superior)
# Ponemos una línea separadora
print("----------------------------------------------------------------")
print("Límite inferior permitido: ",limite_inferior)
```

```python
# Obtenemos datos y los outliers se convierten en nulos en el df
dfDe = cuantitativas[(y <= limite_superior) & (y >= limite_inferior)]
dfDe
```

```python
dfDe.isnull().sum()
```

```python
# Reemplazamos los nulos del df con la media
dfC = dfDe.copy()
dfC.fillna(round(dfDe.mean(), 1), inplace=True)
dfC
```

```python
dfC.isnull().sum()
```

```python
# Veamos el diagrama de cajas de este dataframe
fig = plt.figure(figsize=(30, 30))
dfC.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

## Procedimiento "cuantiles" para sustituir outliers en df

```python
# Método aplicando cuartiles. Encuentro cuartiles 0.25 y 0.75
y = cuantitativas
percentil25 = y.quantile(0.25) # Q1
percentil75 = y.quantile(0.75) # Q3
iqr = percentil75 - percentil25

limite_superior_iqr = percentil75 + 1.5 * iqr
limite_inferior_iqr = percentil25 - 1.5 * iqr
print("Límite superior permitido: ",limite_superior)
# Ponemos una línea separadora
print("----------------------------------------------------------------")
print("Límite inferior permitido: ",limite_inferior)
```

```python
# Obtenemos datos y los outliers se convierten en nulos en el df
df2 = cuantitativas[(y <= limite_superior_iqr) & (y >= limite_inferior_iqr)]
df2
```

```python
df2.isnull().sum()
```

```python
# Reemplazamos los nulos del df con la media
dfiqr = df2.copy()
dfiqr.fillna(round(df2.mean(), 1), inplace=True)
dfiqr
```

```python
fig = plt.figure(figsize=(30, 30))
dfiqr.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

```python
# Unimos el df cuantitativo limpio con el df cualitativo
datos_limpio = pd.concat([cualitativas, dfiqr], axis=1)
datos_limpio
```

```python
datos_limpio.isnull().sum()
```

```python
datos_limpio.to_csv('detalles_clean.csv', index=False)
```
