```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('listings - CDMX.csv')
df
```

```python
# Delimitamos las columnas que seleccionaremos
df = df[['id', 'source', 'name', 'description', 'neighborhood_overview',
       'host_id', 'host_name', 'host_since', 'host_location', 'host_about',
       'host_response_time', 'host_response_rate', 'host_acceptance_rate',
       'host_is_superhost', 'host_listings_count', 'host_total_listings_count',
       'host_verifications', 'host_identity_verified',
       'neighbourhood_cleansed', 'latitude', 'longitude', 'property_type',
       'room_type', 'accommodates', 'bathrooms_text', 'bedrooms', 'beds',
       'amenities', 'price', 'minimum_nights', 'maximum_nights',
       'has_availability', 'availability_30', 'availability_60',
       'availability_90', 'availability_365', 'number_of_reviews',
       'number_of_reviews_ltm', 'number_of_reviews_l30d', 'first_review',
       'last_review', 'review_scores_rating', 'review_scores_accuracy',
       'review_scores_cleanliness', 'review_scores_checkin',
       'review_scores_communication', 'review_scores_location',
       'review_scores_value', 'instant_bookable', 'reviews_per_month']]
df
```

```python
df.isnull().sum()
```

## Tratamiento de valores nulos

```python
# Haremos algo parecido con la columna de description
df['description'] = df['description'].fillna('No description')
df.isnull().sum()
```

```python
# En la columna neighborhood_overview haremos lo mismo
df['neighborhood_overview'] = df['neighborhood_overview'].fillna('No neighborhood_overview')
df.isnull().sum()
```

```python
# Lo mismo con la columna host_location
df['host_location'] = df['host_location'].fillna('No host_location')
df.isnull().sum()
```

```python
# Con host_about, hacemos lo mismo
df['host_about'] = df['host_about'].fillna('No description')
df.isnull().sum()
```

```python
# Con host_response_time, sustituimos los nulos por 'No data'
df['host_response_time'] = df['host_response_time'].fillna('No data')
df.isnull().sum()
```

```python
# Lo mismo con host_response_rate
df['host_response_rate'] = df['host_response_rate'].fillna('No data')
df.isnull().sum()
```

```python
# Al iguall con host_acceptance_rate
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna('No data')
df.isnull().sum()
```

```python
# En host_is_superhost, como es un booleano, lo sustituimos por False por defecto
df['host_is_superhost'] = df['host_is_superhost'].fillna("f")
df.isnull().sum()
```

```python
# Con bathrooms_text, sustituimos los nulos por '1 bath'
df['bathrooms_text'] = df['bathrooms_text'].fillna('1 bath')
df.isnull().sum()
```

```python
# Con bedrooms, sustituimos los nulos por 1 por defecto
df['bedrooms'] = df['bedrooms'].fillna(1)
df.isnull().sum()
```

```python
# Con beds, sustituimos los nulos por 1 por defecto
df['beds'] = df['beds'].fillna(1)
df.isnull().sum()
```

```python
# Con price, primero convertimos los valores a numéricos y después sustituimos los nulos con interpolación
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
df['price'] = pd.to_numeric(df['price'])
df['price'] = df['price'].interpolate()
df.isnull().sum()
```

```python
# Con has_availability, sustituimos los nulos por 'f' por defecto
df['has_availability'] = df['has_availability'].fillna('f')
df.isnull().sum()
```

```python
# Con first_review y last_review, sustituimos los nulos por 'No date'
df['first_review'] = df['first_review'].fillna('No date')
df['last_review'] = df['last_review'].fillna('No date')
df.isnull().sum()
```

```python
# En review_scores_rating, review_scores_accuracy, review_scores_cleanliness, review_scores_checkin, review_scores_communication, 
# review_scores_location, review_scores_value, sustituimos los nulos por 0
df['review_scores_rating'] = df['review_scores_rating'].fillna(0)
df['review_scores_accuracy'] = df['review_scores_accuracy'].fillna(0)
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].fillna(0)
df['review_scores_checkin'] = df['review_scores_checkin'].fillna(0)
df['review_scores_communication'] = df['review_scores_communication'].fillna(0)
df['review_scores_location'] = df['review_scores_location'].fillna(0)
df['review_scores_value'] = df['review_scores_value'].fillna(0)
df.isnull().sum()
```

```python
# En reviews_per_month, sustituimos los nulos con la interpolación
df['reviews_per_month'] = df['reviews_per_month'].interpolate()
# Quitamos el único valor nulo restante
df = df.dropna()
df.isnull().sum()
```

## Valores atípicos

```python
df.info()
```

```python
# Dividimos el dataset en cualitatitvas y cuantitativas
cualitativas = df.select_dtypes(include=['object'])
cuantitativas = df.select_dtypes(include=['float64', 'int64'])
# Dado que los id y las coordenadas no son variables que nos interese analizar cuantitativamente, las quitamos de cuantitativas y lo pasamos a cualitativas
# Procuramos que estas variables estén al incio del dataset
cualitativas = pd.concat([df['id'], df['host_id'], cualitativas], axis=1)
cuantitativas = cuantitativas.drop(['id', 'host_id', 'latitude', 'longitude'], axis=1)
```

```python
# Realizamos un diagrama de caja o bigote de cada columna del df
fig = plt.figure(figsize=(30, 30))
cuantitativas.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

## Procedimiento "desviación estándar" para eliminar outliers en df

```python
# Método aplicando la desviación estándar. Encuentro los valores extremos
# Se considera un valor extremo si está a más de 3 desviaciones estándar de la media
y = cuantitativas
limite_superior = y.mean() + 3 * y.std()
limite_inferior = y.mean() - 3 * y.std()
print("Límite superior permitido: ",limite_superior)
# Ponemos una línea separadora
print("----------------------------------------------------------------")
print("Límite inferior permitido: ", limite_inferior)
```

```python
# Obtenemos datos y los outliers se convierten en nulos en el df
df2 = cuantitativas[(y <= limite_superior) & (y >= limite_inferior)]
df2
```

```python
df2.isnull().sum()
```

```python
# Reemplazamos los nulos del df con la media
dfC = df2.copy()
dfC.fillna(round(df2.mean(), 1), inplace=True)
dfC.isnull().sum()
```

```python
# Realizamos un diagrama de caja o bigote de cada columna del df
fig = plt.figure(figsize=(30, 30))
dfC.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

## Procedimiento "cuantiles" para sustituir outliers en df

```python
# Método aplicando cuartiles. Encuentro cuartiles 0.25 y 0.75
y = cuantitativas
percentil25 = y.quantile(0.25) # Q1
percentil75 = y.quantile(0.75) # Q3
iqr = percentil75 - percentil25

limite_superior_iqr = percentil75 + 1.5 * iqr
limite_inferior_iqr = percentil25 - 1.5 * iqr
print("Límite superior permitido: ",limite_superior)
# Ponemos una línea separadora
print("----------------------------------------------------------------")
print("Límite inferior permitido: ",limite_inferior)
```

```python
# Obtenemos datos y los outliers se convierten en nulos en el df
df3 = cuantitativas[(y <= limite_superior_iqr) & (y >= limite_inferior_iqr)]
df3.isnull().sum()
```

```python
# Reemplazamos los nulos del df con la media
df_iqr = df3.copy()
df_iqr.fillna(round(df3.mean(), 1), inplace=True)
df_iqr
```

```python
fig = plt.figure(figsize=(30, 30))
df_iqr.plot(kind='box', vert=False)
plt.title('Valores atípicos de las variables cuantitativas')
plt.show()
```

```python
# Unimos el df cuantitativo limpio con el df cualitativo
datos_limpio = pd.concat([cualitativas, df_iqr], axis=1)
datos_limpio.isnull().sum()
```

```python
# Pasamos a un excel
datos_limpio.to_csv('listings - CDMX_clean.csv')
```
